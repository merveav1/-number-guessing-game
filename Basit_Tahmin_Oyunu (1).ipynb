```python
# Basit tahmin oyunu: 


import random 

number1=int(input("ilk rakamı giriniz"))
number2=int(input("ikinci rakamı giriniz"))

if number1<number2:
    prediction=int(input("senin tahminin ne?"))
    rastgele=random.randint(number1,number2)
    if prediction==rastgele:
        print("doğru tahmin ettin,kazandın")
    elif not (prediction >= number1 and prediction<=number2):
        print("Tahmin edilen değer iki rakam arasında değil.")
        
    else:
        print(f"yanlış tahmin ettin,kaybettin.Bilgisayarın tahmini:{rastgele}")
else:
    print("ilk rakamı büyük giremezsiniz")



# Ödev için number1 ile number2 arasında tahmin edilen değer girilmez ise
# uyarı yapılacak.
```

```python
# WHILE DÖNGÜSÜ İLE Basit tahmin oyunu: 


import random 

while True:
    
    number1=int(input("ilk rakamı giriniz"))
    number2=int(input("ikinci rakamı giriniz"))

    if number1<number2:
        prediction=int(input("senin tahminin ne?"))
        rastgele=random.randint(number1,number2)
        if prediction==rastgele:
            print("doğru tahmin ettin,kazandın")
        elif not (prediction >= number1 and prediction<=number2):
            print("Tahmin edilen değer iki rakam arasında değil.")

        else:
            print(f"yanlış tahmin ettin,kaybettin.Bilgisayarın tahmini:{rastgele}")
    else:
        print("ilk rakamı büyük giremezsiniz")
```

```python

import random 

print("""
Programdan çıkmak için -99 yazınız.
""")

number1=int(input("ilk rakamı giriniz"))
number2=int(input("ikinci rakamı giriniz"))

dogru = 0
yanlis = 0
    
while True:

    if number1<number2:
        prediction=int(input("senin tahminin ne?"))
        rastgele=random.randint(number1,number2)
        if prediction==rastgele:
            print("doğru tahmin ettin,kazandın")
            dogru += 1
            
        elif prediction==-99:
            print("Programdan çıkılıyor.")
            break
        elif not (prediction >= number1 and prediction<=number2):
            print("Tahmin edilen değer iki rakam arasında değil.")
        
        else:
            print(f"yanlış tahmin ettin,kaybettin.Bilgisayarın tahmini:{rastgele}")
            yanlis +=1
    else:
        print("ilk rakamı büyük giremezsiniz")
        
print(f"Doğru tahmin sayısı {dogru}, yanlış tahmin sayısı {yanlis}")      
        
```
